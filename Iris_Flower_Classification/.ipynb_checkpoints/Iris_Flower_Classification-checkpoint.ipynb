### Label Encoding

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
df['Species'] = encoder.fit_transform(df['Species'])
df.head()
```

### Model Training

```python
from sklearn.model_selection import cross_val_score, train_test_split #to split the dataset for training and testing
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report #for checking the model accuracy
```

```python
# Separate Features and Target
X = df.drop(columns=['Id', 'Species'], axis=1)
y = df['Species']
```

```python
# Train Test Split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=42)
print(X_train.shape, X_test.shape)
print(y_train.shape, y_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression  # for Logistic Regression algorithm

# Model Training
model = LogisticRegression()
model.fit(X_train, y_train)
pred = model.predict(X_test)

# Accuracy of the model
accuracy = accuracy_score(y_test, pred)
print("Accuracy score:", accuracy)
```

```python
from sklearn.svm import SVC  #for Support Vector Machine (SVM) Algorithm

# Model Training
model = SVC()
model.fit(X_train, y_train)
pred = model.predict(X_test)

# Accuracy of the model
accuracy = accuracy_score(y_test, pred)
print("Accuracy score:", accuracy)
```

```python
from sklearn.neighbors import KNeighborsClassifier  # for K nearest neighbours

# Model Training
model = KNeighborsClassifier()
model.fit(X_train, y_train)
pred = model.predict(X_test)

# Accuracy of the model
accuracy = accuracy_score(y_test, pred)
print("Accuracy score:", accuracy)
```

```python
from sklearn.tree import DecisionTreeClassifier #for using Decision Tree Algoithm

# Model Training
model = DecisionTreeClassifier()
model.fit(X_train, y_train)
pred = model.predict(X_test)

# Accuracy of the model
accuracy = accuracy_score(y_test, pred)
print("Accuracy score:", accuracy)
```
